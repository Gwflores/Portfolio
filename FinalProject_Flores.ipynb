```python
position = ['WR', 'CB', 'RB']
drafted = df[(df.Year >= 2009) & (df.Year <=2013) & (df.Drafted == 'Yes') & (df.Position.isin(position))]
drafted = drafted.drop(columns=['Player', 'Drafted..tm.rnd.yr.', 'Position_Type'])
drafted.reset_index()
```

```python
notdrafted = df[(df.Year >= 2014) & (df.Year <=2019) & (df.Drafted == 'No') & (df.Position.isin(position))]
notdrafted = notdrafted.drop(columns=['Player', 'Drafted..tm.rnd.yr.', 'Position_Type'])
notdrafted.reset_index()
```

```python
# Fill na values with Median
drafted['Age'] = drafted['Age'].fillna(drafted['Age'].median())
drafted['Sprint_40yd'] = drafted['Sprint_40yd'].fillna(drafted['Sprint_40yd'].median())
drafted['Vertical_Jump'] = drafted['Vertical_Jump'].fillna(drafted['Vertical_Jump'].median())
drafted['Bench_Press_Reps'] = drafted['Bench_Press_Reps'].fillna(drafted['Bench_Press_Reps'].median())
drafted['Broad_Jump'] = drafted['Broad_Jump'].fillna(drafted['Broad_Jump'].median())
drafted['Agility_3cone'] = drafted['Agility_3cone'].fillna(drafted['Agility_3cone'].median())
drafted['Shuttle'] = drafted['Shuttle'].fillna(drafted['Shuttle'].median())
drafted
```

```python
notdrafted['Age'] = notdrafted['Age'].fillna(notdrafted['Age'].median())
notdrafted['Sprint_40yd'] = notdrafted['Sprint_40yd'].fillna(notdrafted['Sprint_40yd'].median())
notdrafted['Vertical_Jump'] = notdrafted['Vertical_Jump'].fillna(notdrafted['Vertical_Jump'].median())
notdrafted['Bench_Press_Reps'] = notdrafted['Bench_Press_Reps'].fillna(notdrafted['Bench_Press_Reps'].median())
notdrafted['Broad_Jump'] = notdrafted['Broad_Jump'].fillna(notdrafted['Broad_Jump'].median())
notdrafted['Agility_3cone'] = notdrafted['Agility_3cone'].fillna(notdrafted['Agility_3cone'].median())
notdrafted['Shuttle'] = notdrafted['Shuttle'].fillna(notdrafted['Shuttle'].median())
notdrafted
```
